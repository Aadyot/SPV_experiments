```python
k = {}
results = {}
v0s = ["0.10", "0.15", "0.20", "0.25", "0.30", "0.35", "0.40", "0.45"]
for v0 in v0s:
    k[v0] = 6.666/2
    disps = (get_all_data_npy(rf"Data_small_v0\Data\coordinate_p0=3.80_v0={v0}_pin=0_set_"))

    data = get_all_Fs(disps, k[v0], 10)
    timestamps =  data[1][0]
    
    results[v0] = {}

    fig, axs = plt.subplots(2,3, figsize=(15,8))
    # fig.title(f"{v0}")
    mean_Fs = np.zeros_like(data[1][-1])
    max_Fs = np.zeros_like(data[1][-1])
    min_Fs = np.zeros_like(data[1][-1]) + 2
    # subplot 1: abs(Fs) per set
    for i in data.keys():
        axs[0][0].plot(timestamps, np.abs(data[i][-1]))
        mean_Fs += np.abs(data[i][-1])
        max_Fs = np.maximum(max_Fs, np.abs(data[i][-1]))
        min_Fs = np.minimum(min_Fs, np.abs(data[i][-1]))
    mean_Fs /= len(data.keys())
    axs[0][0].set_xscale('log')
    axs[0][0].set_title('abs(Fs) per set')
    results[v0]['Fs'] = mean_Fs
    # subplot 2: MSD per set
    
    axs[1][0].plot(timestamps, max_Fs)
    axs[1][0].plot(timestamps, min_Fs)
    axs[1][0].set_title('Envelope of abs(Fs) across sets')
    axs[1][0].set_xscale('log')
    
    axs[1][2].plot(timestamps, max_Fs - min_Fs)
    axs[1][2].set_title('Range of abs(Fs) across sets')
    axs[1][2].set_xscale('log')
    
    mean_MSD = np.zeros_like(MSD(data[1][1]))
    for i in data.keys():
        axs[0][1].plot(timestamps, MSD(data[i][1]))
        mean_MSD += MSD(data[i][1])
    mean_MSD /= len(data.keys())
    axs[0][1].set_xscale('log')
    axs[0][1].set_yscale('log')
    axs[0][1].set_title('MSD per set')
    results[v0]['MSD'] = mean_MSD

    # subplot 3: X4 (ensemble)
    X4 = X4_calc(data)
    axs[0][2].plot(timestamps, np.abs(X4), label = v0)
    axs[0][2].set_title('abs(X4)')
    axs[0][2].set_xscale('log')
    results[v0]['X4'] = X4

    fig.tight_layout()
    plt.show()
    
```

```python
for v0 in v0s:
    plt.plot(timestamps, np.abs(results[v0]['X4']), label = v0)
plt.legend()            
plt.xscale('log')
```
