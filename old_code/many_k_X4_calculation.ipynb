```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7, 10), sharex=True)


for k in [np.pi/1.2]:
    print(k)
    data = get_all_Fs(data_0, k)
    timestamps = data[1][0]

    Fs_avg = np.zeros_like(data[1][-1])
    for i in data.keys():
        Fs_avg += data[i][-1]
    ax1.plot(timestamps, np.abs(Fs_avg)/10, label=k)
    ax1.set_xscale('log')
    
    X4 = X4_calc(data)   
    ax2.plot(timestamps, np.abs(X4), label=k)

    ax2.set_xscale('log')

ax1.legend()

ax2.legend()

plt.tight_layout()
plt.show()
```
